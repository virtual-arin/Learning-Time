```python
import torch
x = torch.tensor([1., 2., 3., 4.])
y = torch.tensor([2., 4., 6., 8.])

w = torch.tensor(0.0)
b = torch.tensor(0.0)

y_pred = w * x + b

loss = torch.mean((y_pred - y) ** 2)

print("Actual values:", y)
print("Predicted values:", y_pred)
print("Loss:", loss)
```
