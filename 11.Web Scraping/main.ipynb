```python
import yfinance as yf

# Step 1: Fetch historical stock data for NVIDIA
nvidia = yf.Ticker("NVDA")
nvidia_data = nvidia.history(period="5y")  # Fetch data for the last 5 years

# Step 2: Reset the index to have 'Date' as a column
nvidia_data.reset_index(inplace=True)

# Step 3: Save the data to a CSV file
nvidia_data.to_csv('./data/nvidia_stock_data.csv', index=False)

print(nvidia_data.head())
```

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time

# Step 1: Set up Chrome WebDriver
chrome_driver_path = 'chromedriver.exe'

# Step 2: Create a Service object and initialize the WebDriver
service = Service(chrome_driver_path)

# Step 3: Open the Chrome browser using the WebDriver
driver = webdriver.Chrome(service=service)

# Step 4: Navigate to the URL
url = "https://www.macrotrends.net/stocks/charts/GOOGL/alphabet/revenue"
driver.get(url)

# Step 5: Wait for the page to load
time.sleep(5) 

# Step 6: Locate the table containing the revenue data
table = driver.find_element(By.XPATH, '//table')

# Step 7: Extract the rows from the table
rows = table.find_elements(By.TAG_NAME, 'tr')
data = []

# Step 8: Loop through the rows and extract the year and revenue
for row in rows[1:]:
    cells = row.find_elements(By.TAG_NAME, 'td')
    if len(cells) > 1:
        year = cells[0].text
        revenue = cells[1].text
        data.append([year, revenue])

# Step 9: Create a DataFrame from the extracted data
google_revenue = pd.DataFrame(data, columns=['Year', 'Revenue'])

# Step 10: Save the DataFrame to a CSV file
google_revenue.to_csv('./data/google_revenue.csv', index=False)
print(google_revenue.head())

driver.quit()
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

books_info = []

# 1. Start the loop to iterate through all 50 pages
for i in range(1, 51):
    # Ensure the URL updates with each iteration of 'i'
    website = f"https://books.toscrape.com/catalogue/page-{i}.html"
    
    # 2. Make the network request
    request = requests.get(website)
    soup = BeautifulSoup(request.content, "html.parser")
    
    # 3. Find all book containers on the current page
    books = soup.find_all("li", class_="col-xs-6 col-sm-4 col-md-3 col-lg-3")

    # 4. Loop through each book found on the page
    for item in books:
        tag = item.find("h3").find("a")
        name = tag["title"]  # Use ['title'] because text is often truncated with '...'
        link = tag["href"]
        price = item.find("p", class_="price_color").text.strip()

        # Create a dictionary for the book data
        product = {
            "Name": name,
            "Price": price,
            "Link": link
        }

        # Add the dictionary to our list
        books_info.append(product)
    
    print(f"Finished scraping page {i}")

# 5. Convert the list of dictionaries into a DataFrame and save
df = pd.DataFrame(books_info)
df.to_csv("./data/books_info.csv", index=False)

print("Scraping complete! File saved as books_info.csv")
```
